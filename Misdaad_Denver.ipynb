# Data Understanding and Preperation

Dit document beschrijft een voorbereidende analyse van de data uit “crime.csv”. De data betreft misdaadstatistieken uit Denver, zoals ook al genoemd in de business understanding. Een voorbeeld van hoe de onderstaande data eruit ziet is in de data preparation te vinden

###### INCIDENT_ID 
Unieke identifier (ID) voor elk incident. Als een incident meerdere overtredingen bevat, bijvoorbeeld autodiefstal en ontvoering, dan wordt voor elke overtreding een OFFENSE_ID's aangemaakt met hetzelfde INCIDENT_ID

###### OFFENSE_CODE - 
Unieke identifier (ID) voor elk soort overtreding dat bestaat uit vier cijfers. 

###### OFFENSE_CODE_EXTENSION
Dit wordt gebruikt om een subset te beschrijven van een ander type overtreding. Bijvoorbeeld ‘traf-other’ en ‘vehicular-eluding-no-chase’ hebben dezelfde OFFENSE_CODE.  

###### OFFENSE_ID 
Unieke identifier (ID) voor elke overtreding. Het OFFENSE_ID is een aaneengeschakelde combinatie van het INCIDENT_ID, OFFENSE_CODE en OFFENSE_CODE_EXTENSION. De eerste 4 getallen lijkt op het jaartal, dat overeenkomt met het jaartal in REPORTED DATE. 

###### OFFENSE_TYPE_ID 
De naam van de overtreding, waarbij de naam kan worden herleidt uit de OFFENSE_CODE in combinatie met OFFENSE_EXTENSION. 

###### OFFENSE_CATEGORY_ID 
Overkoepelende categorisatie van overtredingen. Bijvoorbeeld ‘theft-from-bldg’ en ‘theft-other’ hoort beide bij de categorie ‘larceny’ 

###### FIRST_OCCURENCE_DATE 
Eerst mogelijke datum van het incident. Als de tijd van het incident bekend is, zal deze waarde toegevoegd worden aan de datum. 

###### LAST_OCCURENCE_DATE 
Laatst mogelijk datum van het incident. Als de tijd van het incident bekend is, is de waarde ‘NaN’. Deze waarde kan een datum + tijd zijn als alleen de mogelijke tijden bekend zijn waarin het incident heeft plaatsgevonden. 

###### REPORTED_DATE 
De datum en tijd wanneer het incident is gerapporteerd aan de politie. 

###### INCIDENT_ADDRESS 
De locatie van het incident. 

###### GEO_LON / GEO_LAT 
Lengte- en breedtegraad van de locatie van het incident. 

###### GEO_X / GEO_Y 
De locatie van het incident die bepaal is met het STPS (State Plane Coordinate System) https://en.wikipedia.org/wiki/State_Plane_Coordinate_System

###### DISTRICT_ID 
Het district van Denver die het incident heeft verwerkt aangeduid met één cijfer. 

###### PRECINCT_ID 
Het politiebureau dat het incident heeft verwerkt aangeduid met drie cijfers. 

###### NEIGHBORHOOD_ID 
Naam van de buurt waarin het incident heeft plaatsgevonden. 

###### IS_CRIME 
Een waarde van 0 of 1 om aan te geven of de overtreding een misdaad is. 0 is geen misdaad. 1 is wel een misdaad. 

###### IS_TRAFFIC 
Een waarde van 0 of 1 om aan te geven of de overtreding een verkeersongeval was. 0 is geen verkeersongeval. 1 is wel een verkeersongeluk.

```python
df = crime.REPORTED_DATE.value_counts().reset_index()
df.rename(columns={'index' : 'Jaar', 'REPORTED_DATE' : 'Aantal crime'}, inplace=True)
x = df[['Jaar']]
y = df['Aantal crime']
reg = LinearRegression().fit(x, y)
print(reg.score(x, y))
print(reg.coef_)
```

Uit de r2 waarden is te halen dat dit verband zeer zwak is en hier geen goede voorspelling uit te halen

```python
for i in crime.sort_values(by='DISTRICT_ID').DISTRICT_ID.unique():
    district = crime.loc[crime['DISTRICT_ID'] == i]
    district = district.REPORTED_DATE.value_counts().reset_index()
    district.rename(columns={'index' : 'Jaar', 'REPORTED_DATE' : 'Aantal crime'}, inplace=True)
    x = district[['Jaar']]
    y = district['Aantal crime']
    dis = LinearRegression().fit(x, y)
    print("District %s: " %(i))
    print(dis.score(x, y))
    print(dis.coef_,"\n")
```

De r2 waarde van district 1 is 0.06 dit houd in dat er geen liniear verband is en hier ook geen goede voorspelling uit gehaald kan worden. <br>
Voor district 2 is dit al iets beter met 0.45. Alleen is deze waarden nog steeds te laag om een betrouwbare voorspelling te doen.<br>
